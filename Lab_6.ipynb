```python
import random

def random_bottle_weight():
  return 9 if random.random() < 0.1 else 10
result = [random_bottle_weight() for _ in range(1000)]
print(f"9oz bottles: {result.count(9)}")
print(f"10oz bottles: {result.count(10)}")
print(f"Proportion of 9oz: {result.count(9)/1000:.3f}")
```

```python
def expected_weight(num_trails = 1000):
  total_weight = 0
  for _ in range(num_trails):
    total_weight += random_bottle_weight()
    return total_weight / num_trails

average_weight = expected_weight(100000)
print(f"Average weight: {average_weight:.4f}oz")
print(f"Do they agree? {'Yes' if abs(average_weight - 9.9) < 0.01 else 'No'}")
```

```python
def pulls_until_short():
    """Returns number of pulls until finding a 9 oz bottle"""
    pulls = 0
    while True:
        pulls += 1
        if random_bottle_weight() == 9:
            return pulls

def test_pulls_simulation(num_simulations=10000):
    total_pulls = 0
    results = []
    for _ in range(num_simulations):
        pulls = pulls_until_short()
        total_pulls += pulls
        results.append(pulls)

    avg_pulls = total_pulls / num_simulations
    print(f"Average pulls to find short bottle: {avg_pulls:.2f}")
    print(f"Theoretical expected pulls: 10.00")
    print(f"Median pulls: {sorted(results)[len(results)//2]}")
    print(f"Min pulls: {min(results)}, Max pulls: {max(results)}")

    # Check what percentage find it in <= 10 pulls
    within_10 = sum(1 for x in results if x <= 10)
    print(f"Percentage found within 10 pulls: {within_10/len(results)*100:.1f}%")

    return results

# Run the simulation
results = test_pulls_simulation(10000)
```

```python
import random

def create_realistic_bottles():
    bottles = [10] * 100

    short_case_start = random.randint(0, 9) * 10

    for i in range(short_case_start, short_case_start + 10):
        bottles[i] = 9

    return bottles, short_case_start // 10

def realistic_pulls_until_short():
    bottles, short_case = create_realistic_bottles()

    shuffled_bottles = bottles.copy()
    random.shuffle(shuffled_bottles)

    for i, weight in enumerate(shuffled_bottles, 1):
        if weight == 9:
            return i, short_case

    return len(shuffled_bottles), short_case

def improved_simulation(num_simulations=10000):
    total_pulls = 0
    results = []

    for _ in range(num_simulations):
        pulls, short_case = realistic_pulls_until_short()
        total_pulls += pulls
        results.append(pulls)

    avg_pulls = total_pulls / num_simulations
    print(f"Average pulls to find short bottle: {avg_pulls:.2f}")
    print(f"Theoretical expected: 100/11 ≈ 9.09 (since 10 short out of 100)")
    print(f"Median pulls: {sorted(results)[len(results)//2]}")


    return results

improved_results = improved_simulation(5000)
```

```python
class PlateStack:
    def __init__(self):
        self.stack = []

    def push(self, plate):
        self.stack.append(plate)
        print(f"Pushed: {plate}. Current stack: {self.stack}")

    def pop(self):
        if not self.is_empty():
            removed_plate = self.stack.pop()
            print(f"Popped: {removed_plate}. Current stack: {self.stack}")
            return removed_plate
        else:
            print("Stack is empty, cannot pop.")
            return None

    def peek(self):
        if not self.is_empty():
            print(f"Top plate: {self.stack[-1]}")
            return self.stack[-1]
        else:
            print("Stack is empty.")
            return None

    def is_empty(self):
        return len(self.stack) == 0

    def size(self):
        return len(self.stack)

stack_of_plates = PlateStack()
stack_of_plates.push("Red Plate")
stack_of_plates.push("Blue Plate")
stack_of_plates.push("Green Plate")

stack_of_plates.peek()
stack_of_plates.pop()
stack_of_plates.peek()
stack_of_plates.pop()
stack_of_plates.pop()
stack_of_plates.pop()
```

```python
import random

def simulate_plate_drawing_with_rejection(num_simulations=100000):
    cabinet_stack = ['P'] * 6 + ['NP'] * 18
    random.shuffle(cabinet_stack)

    dishwasher = []

    exactly_one_pink_count = 0
    at_least_one_pink_count = 0

    for sim_idx in range(num_simulations):
        if len(dishwasher) >= 12 or len(cabinet_stack) < 4:
            combined_plates = dishwasher + cabinet_stack
            random.shuffle(combined_plates)
            cabinet_stack = combined_plates
            dishwasher = []

        if len(cabinet_stack) < 4:
            print(f"Warning: Not enough plates for a table setting after reshuffle in simulation {sim_idx}. Skipping this iteration.")
            continue

        current_draw_for_table = []

        initial_pull = []
        for _ in range(4):
            initial_pull.append(cabinet_stack.pop())

        if 'P' in initial_pull:
            current_draw_for_table = initial_pull
        else:
            rejected_plates_temp = list(initial_pull)
            additional_pulls_for_rejection = []
            found_pink_plate = None

            while not found_pink_plate:
                if not cabinet_stack:
                    all_plates_in_system = dishwasher + cabinet_stack + rejected_plates_temp + additional_pulls_for_rejection
                    random.shuffle(all_plates_in_system)
                    cabinet_stack = all_plates_in_system
                    dishwasher = []
                    rejected_plates_temp.clear()
                    additional_pulls_for_rejection.clear()

                    if not cabinet_stack:
                        print(f"Critical Error: Cabinet empty after emergency reshuffle in simulation {sim_idx}. Skipping this iteration.")
                        break

                next_plate = cabinet_stack.pop()
                if next_plate == 'P':
                    found_pink_plate = next_plate
                else:
                    additional_pulls_for_rejection.append(next_plate)

            plates_to_return = rejected_plates_temp + additional_pulls_for_rejection
            if found_pink_plate:
                plates_to_return.append(found_pink_plate)

            cabinet_stack.extend(plates_to_return)

            if len(cabinet_stack) < 4:
                all_plates_in_system = dishwasher + cabinet_stack
                random.shuffle(all_plates_in_system)
                cabinet_stack = all_plates_in_system
                dishwasher = []
                if len(cabinet_stack) < 4:
                    print(f"Critical Error: Cabinet empty even after secondary full reshuffle in simulation {sim_idx}. Skipping this iteration.")
                    continue

            for _ in range(4):
                current_draw_for_table.append(cabinet_stack.pop())


        if current_draw_for_table:
            dishwasher.extend(current_draw_for_table)

            pink_plates_in_draw = current_draw_for_table.count('P')

            if pink_plates_in_draw == 1:
                exactly_one_pink_count += 1
            if pink_plates_in_draw >= 1:
                at_least_one_pink_count += 1


    prob_exactly_one_pink = exactly_one_pink_count / num_simulations
    prob_at_least_one_pink = at_least_one_pink_count / num_simulations

    print(f"Simulated probability of exactly one pink plate: {prob_exactly_one_pink:.4f}")
    print(f"Simulated probability of at least one pink plate: {prob_at_least_one_pink:.4f}")

simulate_plate_drawing_with_rejection(100000)
```

```python
import random

def simulate_plate_drawing_with_rejection_and_all_pink(num_simulations=1000):
    pink_plates = [f'P_{i+1}' for i in range(6)]
    non_pink_plates = [f'NP_{i+1}' for i in range(18)]
    cabinet_stack = pink_plates + non_pink_plates
    random.shuffle(cabinet_stack)

    dishwasher = []

    plate_usage_counts = {plate_id: 0 for plate_id in (pink_plates + non_pink_plates)}

    exactly_one_pink_count = 0
    at_least_one_pink_count = 0
    all_pink_plates_count = 0

    def pop_and_track(stack_list):
        if stack_list:
            plate = stack_list.pop()
            plate_usage_counts[plate] += 1
            return plate
        return None

    for sim_idx in range(num_simulations):

        if len(dishwasher) >= 12 or len(cabinet_stack) < 4:
            combined_plates = dishwasher + cabinet_stack
            random.shuffle(combined_plates)
            cabinet_stack = combined_plates
            dishwasher = []

        if len(cabinet_stack) < 4:
            continue

        current_draw_for_table = []

        initial_pull = []
        for _ in range(4):
            plate = pop_and_track(cabinet_stack)
            if plate:
                initial_pull.append(plate)

        if any(plate.startswith('P') for plate in initial_pull):
            current_draw_for_table = initial_pull
        else:
            rejected_plates_temp = list(initial_pull)
            additional_pulls_for_rejection = []
            found_pink_plate = None

            while not found_pink_plate:
                if not cabinet_stack:
                    all_plates_in_system = dishwasher + cabinet_stack + rejected_plates_temp + additional_pulls_for_rejection
                    random.shuffle(all_plates_in_system)
                    cabinet_stack = all_plates_in_system
                    dishwasher = []
                    rejected_plates_temp.clear()
                    additional_pulls_for_rejection.clear()

                    if not cabinet_stack:
                        break

                next_plate = pop_and_track(cabinet_stack)
                if next_plate is None:
                    break

                if next_plate.startswith('P'):
                    found_pink_plate = next_plate
                else:
                    additional_pulls_for_rejection.append(next_plate)

            plates_to_return = rejected_plates_temp + additional_pulls_for_rejection
            if found_pink_plate:
                plates_to_return.append(found_pink_plate)

            cabinet_stack.extend(plates_to_return)

            if len(cabinet_stack) < 4:
                all_plates_in_system = dishwasher + cabinet_stack
                random.shuffle(all_plates_in_system)
                cabinet_stack = all_plates_in_system
                dishwasher = []
                if len(cabinet_stack) < 4:
                    continue

            for _ in range(4):
                plate = pop_and_track(cabinet_stack)
                if plate:
                    current_draw_for_table.append(plate)


        if current_draw_for_table and len(current_draw_for_table) == 4:
            dishwasher.extend(current_draw_for_table)

            pink_plates_in_draw = sum(1 for plate in current_draw_for_table if plate.startswith('P'))

            if pink_plates_in_draw == 1:
                exactly_one_pink_count += 1
            if pink_plates_in_draw >= 1:
                at_least_one_pink_count += 1
            if pink_plates_in_draw == 4:
                all_pink_plates_count += 1

    prob_exactly_one_pink = exactly_one_pink_count / num_simulations
    prob_at_least_one_pink = at_least_one_pink_count / num_simulations
    prob_all_pink_plates = all_pink_plates_count / num_simulations

    print(f"Number of simulations (table setting events): {num_simulations}")
    print(f"Events with exactly one pink plate: {exactly_one_pink_count}")
    print(f"Simulated probability of exactly one pink plate: {prob_exactly_one_pink:.4f}")
    print(f"Events with at least one pink plate: {at_least_one_pink_count}")
    print(f"Simulated probability of at least one pink plate: {prob_at_least_one_pink:.4f}")
    print(f"Events with all four pink plates: {all_pink_plates_count}")
    print(f"Simulated probability of all four pink plates: {prob_all_pink_plates:.4f}")
    print("\nPlate Usage Counts:")
    sorted_plate_usage = dict(sorted(plate_usage_counts.items(), key=lambda item: (item[0].split('_')[0], int(item[0].split('_')[1]))))
    for plate_id, count in sorted_plate_usage.items():
        print(f"  {plate_id}: {count} draws")

simulate_plate_drawing_with_rejection_and_all_pink(1000)
```
