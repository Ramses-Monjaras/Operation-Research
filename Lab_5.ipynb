#Recursion Assignment

##Analytic Eqation

```python
import numpy as np
import time
from functools import lru_cache

def analytic_f(n):
    return (2/3) * (2**n) + (1/3) * (5**n)
for i in range(10):
    print(f"f({i}) = {analytic_f(i)}")
```

##Dynamic Programming

```python
def dp_f(n):
    if n == 0:
        return 1
    if n == 1:
        return 3

    dp = [0] * (n + 1)
    dp[0] = 1
    dp[1] = 3

    for i in range(2, n + 1):
        dp[i] = 7 * dp[i-1] - 10 * dp[i-2]

    return dp[n]
for i in range(10):
    print(f"f({i}) = {dp_f(i)}")
```

##Memoization

```python
def memo_f(n):
    if n == 0:
        return 1
    if n == 1:
        return 3

    return 7 * memo_f(n-1) - 10 * memo_f(n-2)
for i in range(10):
    print(f"f({i}) = {memo_f(i)}")
```

It seems that each function got the same anwser. The Analytic Solution seems to work the fastest for large n, but it requires mathematical derivation. Dynamic Programming seems to have a good balance of performance, but it is slow when computing large numbers. Memoization seems to be more intuituve, but has recursion limits.
